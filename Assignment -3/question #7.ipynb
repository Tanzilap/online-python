```python
#Write a Python program which take 5 inputs from user for different subjects marks,total it and generate mark sheet using grades?
Eng=int(input("Enter marks="));
Math=int(input("Enter marks="));
Urdu=int(input("Enter marks="));
Isl=int(input("Enter marks="));
Sci=int(input("Enter marks="));
Totalmarks=500;
print("Total marks is:",Totalmarks)
Obtained=Eng+Math+Urdu+Isl+Sci;
print("Obtained Marks is:",Obtained)
P=(Obtained/Totalmarks)*100;
print("Percentage=",P,"%")
if P < 100 and P >= 80:
    print("Grade A+");
elif P < 80 and P >= 70:
    print("Grade A");
elif P < 70 and P >= 60:
    print("Grade B");
elif P < 60 and P >= 50:
    print("Grade C");
else:
    print("Fail")
```
