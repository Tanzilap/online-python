```python
# Make a calculator using Python with addition , subtraction , multiplication ,division and power.
a=int(input("enter your num:"))
b=int(input("enter your num:"))
c=input("enter your operator:")
if c == "+":
    sum=(a+b)
    print("sum:",sum)
elif c == '-':
    sub=(a-b)
    print("sub:",sub)
elif c == '*':
        multiply=(a+b)
        print("multiply",multiply)
elif c =='/':
            divide=(a/b)
            print("divide",divide)
elif c == '**':
    power=a**b
    print("power",power)
    else:
        print("operator is not defined")
```
